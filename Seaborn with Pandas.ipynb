```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = 'Data/Sample Data/mpg_report.csv'
data_set = pd.read_csv(file)
```

```python
data_set.columns
```

```python
data_set.head()
```

```python
data_set['origin'].unique()
```

```python
sns.relplot(x='horsepower', y='displacement', data=data_set, kind='scatter')
```

```python
sns.relplot(x='horsepower', y='displacement', data=data_set, kind='scatter', hue='acceleration', size='weight')
```

```python
s = sns.relplot(y='displacement', x='horsepower', data=data_set, kind='scatter', size="cylinders", style="cylinders", hue='origin', alpha=0.9)
s.fig.set_figwidth(10)
s.fig.set_figheight(7)
```
